# Concat y Merge en Pandas

```python
import pandas as pd
```

Crearemos algunos DataFrames de ejemplo para usar en nuestros ejemplos.

```python
df1 = pd.DataFrame({'A': ['A0', 'A1', 'A2', 'A3'],
                    'B': ['B0', 'B1', 'B2', 'B3'],
                    'C': ['C0', 'C1', 'C2', 'C3'],
                    'D': ['D0', 'D1', 'D2', 'D3']})

df2 = pd.DataFrame({'A': ['A4', 'A5', 'A6', 'A7'],
                    'B': ['B4', 'B5', 'B6', 'B7'],
                    'C': ['C4', 'C5', 'C6', 'C7'],
                    'D': ['D4', 'D5', 'D6', 'D7']})

df3 = pd.DataFrame({'A': ['A8', 'A9', 'A10', 'A11'],
                    'B': ['B8', 'B9', 'B10', 'B11'],
                    'C': ['C8', 'C9', 'C10', 'C11'],
                    'D': ['D8', 'D9', 'D10', 'D11']})

df4 = pd.DataFrame({'A': ['A0', 'A1', 'A2', 'A3'],
                    'B': ['B0', 'B1', 'B2', 'B3'],
                    'C': ['C0', 'C1', 'C2', 'C3'],
                    'E': ['E0', 'E1', 'E2', 'E3']})
```

## Concatenación con concat
Ahora podemos concatenar los DataFrames verticalmente (es decir, uno debajo del otro) utilizando concat() con axis=0:

```python
vertical_concatenation = pd.concat([df1, df2, df3], axis=0)
vertical_concatenation.reset_index(drop=True)
```

```python
pd.concat([df1, df4], axis=0)
```

También podemos concatenar los DataFrames horizontalmente (es decir, uno al lado del otro) utilizando concat() con axis=1:

```python
horizontal_concatenation = pd.concat([df1, df2, df3], axis=1)
horizontal_concatenation
```

## Merge

```python
df1 = pd.DataFrame({'key': ['K3', 'K1', 'K2', 'K0'],
                    'A': ['A3', 'A1', 'A2', 'A0'],
                    'B': ['B3', 'B1', 'B2', 'B0']})

df2 = pd.DataFrame({'key': ['K0', 'K1', 'K2', 'K4'],
                    'C': ['C0', 'C1', 'C2', 'C3'],
                    'D': ['D0', 'D1', 'D2', 'D3']})
```

```python
df1
```

```python
df2
```

Podemos unir estos DataFrames usando la columna 'key' que tienen en común:

```python
merged_df = pd.merge(df1, df2, on='key')
merged_df
```

```python
pd.merge(df1, df2, on='key', how='left')
```

```python
pd.merge(df1, df2, on='key', how='right')
```

```python
pd.merge(df1, df2, how='cross')
```

También podemos unir los DataFrames utilizando múltiples columnas en común:

```python
df3 = pd.DataFrame({'key1': ['K0', 'K0', 'K1', 'K2'],
                    'key2': ['K0', 'K1', 'K0', 'K1'],
                    'E': ['E0', 'E1', 'E2', 'E3'],
                    'F': ['F0', 'F1', 'F2', 'F3']})

df4 = pd.DataFrame({'key1': ['K0', 'K1', 'K1', 'K2'],
                    'key2': ['K0', 'K0', 'K0', 'K0'],
                    'G': ['G0', 'G1', 'G2', 'G3'],
                    'H': ['H0', 'H1', 'H2', 'H3']})
```

```python
df3
```

```python
df4
```

```python
pd.merge(df3, df4, on=['key1', 'key2'], how="left")
```
